## **Bag of Elements and Concatenation of the Vector**

```python
# List to store unique elements from 'Reactant_1' column
unique_elements_reactant_1 = []

# Iterate through 'Reactant_1' column to extract unique elements
for smiles in df['Reactant_1']:
    # Check if the entry is a valid SMILES string
    if Chem.MolFromSmiles(smiles) is not None:
        mol = Chem.MolFromSmiles(smiles)
        atoms = [atom.GetSymbol() for atom in mol.GetAtoms()]
        unique_elements_reactant_1.extend(atoms)

# Get unique elements as a list
unique_elements_list_reactant_1 = list(set(unique_elements_reactant_1))

print("Unique elements in the 'Reactant_1' column as a list:")
print(unique_elements_list_reactant_1)
```

```python
all_lists = [unique_elements_list_reactant_1,
             unique_elements_list_reactant_2,
             unique_elements_list_ligand,
             unique_elements_list_reagent,
             unique_elements_list_solvent]

unique_elements_set = set() # Initialize an empty set for unique elements

# Iterate through each list and add elements to the set
for lst in all_lists:
    unique_elements_set.update(lst)

unique_elements_combined = list(unique_elements_set)  # Convert set to list

print("Unique elements in the dataset:")
print(unique_elements_combined)

```

```python
#Define a function for Bag of Elements vector
def smiles_to_bagofelements(smiles_string):

  #Create a numpy array for bag-of-elements vector
  bag_of_elements = np.zeros(len(unique_elements_combined), dtype = int)

  #Iterate through search elements list with the following loop
  for i, element in enumerate(unique_elements_combined):
    while element in smiles_string:

      #If the element is in the SMILES string add 1 to the bag_of_elements string
      bag_of_elements[i] += 1

      #Find the index of the corresponding element in the SMILES string
      index = smiles_string.index(element)

      #Use the index variable you have defined to remove that element from the SMILES string
      smiles_string = smiles_string[:index] + smiles_string[index + len(element):]

  return tuple(bag_of_elements)

df['BoE_Reactant_1'] = df['Reactant_1'].apply(smiles_to_bagofelements)

boe_df = df[['BoE_Reactant_1']]
```

## **Classification**

```python
!pip install rdkit-pypi
import rdkit
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy.linalg import norm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from rdkit import DataStructs
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from scipy import stats
from scipy.stats import zscore
from sklearn.ensemble import RandomForestRegressor
from sklearn.utils import resample
from numpy import linalg as LA
from sklearn.preprocessing import OneHotEncoder
```

```python
#Set features and target axis for model to train
features = boe_df['concatenated_vector']

target = boe_df['class_success']

#Split it by 80/20
# Train test split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=0)

# Reshape the features to 2D arrays
X_train = np.vstack(X_train.values)
X_test = np.vstack(X_test.values)
```

```python
#Logistic Regression
clf = LogisticRegression(random_state = 0)
scores = cross_val_score(clf, X_train, y_train, cv=5)
Logistic_score = np.mean(scores)
print('Logistic Regression CV score: ' + str(np.mean(scores)))

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

Logistic_score_test = accuracy_score(y_test, y_pred)
print('Logisitic regression test score: ' + str(Logistic_score_test))
```

```python
print('Logistic Regression CV score: ' + str(np.mean(scores)))
print('Logisitic regression test score: ' + str(Logistic_score_test))
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
scores = cross_val_score(knn, X_train, y_train, cv=5)
knn_score = np.mean(scores)
print('knn CV score: ' + str(np.mean(scores)))

knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

knn_score_test = accuracy_score(y_test, y_pred)
print('KNN test score: ' + str(knn_score_test))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
scores = cross_val_score(rf, X_train, y_train, cv=5)
rf_score = np.mean(scores)
print('Random Forest CV score: ' + str(np.mean(scores)))

rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

rf_score_test = accuracy_score(y_test, y_pred)
print('Random Forest test score: ' + str(rf_score_test))
```

```python
#Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier

gbc = GradientBoostingClassifier()
scores = cross_val_score(gbc, X_train, y_train, cv=5)
gbc_score = np.mean(scores)
print('Gradient Boosting Classifier CV score: ' + str(np.mean(scores)))

gbc.fit(X_train, y_train)
y_pred = gbc.predict(X_test)

gbc_score_test = accuracy_score(y_test, y_pred)
print('Gradient Boosting Classifier test score: ' + str(gbc_score_test))
```

```python
#Gaussian MB
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
scores = cross_val_score(gnb, X_train, y_train, cv=5)
nb_score = np.mean(scores)
print('Gaussian NB CV score: ' + str(np.mean(scores)))

gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)

gnb_score_test = accuracy_score(y_test, y_pred)
print('Gaussian NB test score: ' + str(gnb_score_test))
```

```python
#Support Vector
from sklearn.svm import SVC

svc = SVC()
scores = cross_val_score(svc, X_train, y_train, cv=5)
svc_score = np.mean(scores)
print('Support vector classifier CV score: ' + str(np.mean(scores)))

svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)

svc_score_test = accuracy_score(y_test, y_pred)
print('Support vector classifier test score: ' + str(svc_score_test))
```

```python
models = ['Logistic Regression','KNN','Random Forest','Gradient Boosting','GaussianNB','Support Vector']
CV_scores = [Logistic_score,knn_score,rf_score,gbc_score,nb_score,svc_score]
test_scores = [Logistic_score_test,knn_score_test,rf_score_test,gbc_score_test,gnb_score_test,svc_score_test]

plt.figure(figsize=(15,8))
plt.bar(models, CV_scores, color='blue', alpha=0.7, label='CV Accuracy', width=0.4)
plt.bar(np.arange(len(models)) + 0.4, test_scores, color='orange', alpha=0.7, label='Test Accuracy', width=0.4)
plt.title('Accuracy scores for different models')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
