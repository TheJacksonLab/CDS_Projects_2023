```python
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

xlabels =['Predicted Success', 'Predicted Fail']
ylabels =['Actual Success', 'Actual Fail']

array = [[578,112],[41,421]]
ax = sn.heatmap(data = array,xticklabels=xlabels,yticklabels=ylabels,fmt='g', annot=True,annot_kws={"size": 20})
plt.title('NN using bag-of-elements')
plt.show()
```
